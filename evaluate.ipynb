# Output Evaluation

```python
import numpy as np
import pandas as pd
from torchvision import transforms
import os
import PIL
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def readFile(file, new_file_path):
    output = []
    with open(file) as f:
        lines = f.readlines()
    lines = np.array(lines)
    i = 0
    while i < lines.shape[0]:
        list = []
        # filepath
        list.append(lines[i][12:-1])
        # true value
        list.append(lines[i+1][7:-1])
        # model predicted value
        list.append(lines[i+2][12:-1])

        output.append(list)
        # jump to the next info
        i += 6

    output = np.array(output)

    for data in output:
        filepath = data[0]
        if filepath[0:3] == '../':
            filepath = filepath[3:]
        # open the image file
        image = PIL.Image.open(filepath)
        new_dir = new_file_path + data[1] + '_to_'+data[2] + '/'
        if not os.path.exists(new_dir):
            os.mkdir(new_dir)
        new_path = new_file_path + data[1] + '_to_'+data[2] + '/' + filepath[-12:]

        # store the file 
        image.save(new_path)
```

```python
file_path = "march_models/soybean_model_resize/b1/test_result.txt"
dir_path = "march_models/soybean_model_resize/b1/comparison/"
readFile(file_path, dir_path)
```

```python
file_path = "march_models/corn_model_resize/nowd/test_result.txt"
dir_path = "march_models/corn_model_resize/nowd/comparison/"
readFile(file_path, dir_path)
```
